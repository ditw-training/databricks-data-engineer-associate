## Task 2: Write a `databricks.yml` Bundle Root Config

Complete the `databricks.yml` file for a bundle called `orders_pipeline`.

Requirements:
- Bundle name: `orders_pipeline`
- Workspace host: `https://adb-1234567890.azuredatabricks.net`
- Two targets: `dev` (mode: development) and `prod` (mode: production)
- Variable `catalog`: `dev_catalog` in dev, `prod_catalog` in prod

**TODO:** Fill in the blanks:
```yaml
bundle:
  name: # TODO: your_bundle_name

workspace:
  host: # TODO: https://adb-<workspace-id>.azuredatabricks.net

targets:
  dev:
    mode: # TODO: development
    variables:
      catalog: # TODO: dev_catalog_name
  prod:
    mode: # TODO: production
    variables:
      catalog: # TODO: prod_catalog_name
```

````python
# Task 2: Complete solution
solution = '''
bundle:
  name: orders_pipeline

workspace:
  host: https://adb-1234567890.azuredatabricks.net

targets:
  dev:
    mode: development
    variables:
      catalog: dev_catalog
  prod:
    mode: production
    variables:
      catalog: prod_catalog
'''
print('Task 2 solution:'); print(solution)
````

## Task 3: Add a Notebook Task to the Bundle

Add a `resources/jobs/ingest_orders.yml` resource file for a job that:
- Job name: `ingest_orders_job`
- One notebook task: `task_key: bronze_load`
- Notebook path: `/Workspace/Users/<user>/orders_pipeline/notebooks/bronze_load`
- Source: `WORKSPACE`
- Uses the `Default` environment

**TODO:** Complete the YAML:
```yaml
resources:
  jobs:
    ingest_orders_job:
      name: # TODO: job_name
      tasks:
        - task_key: # TODO: task_key_name
          notebook_task:
            notebook_path: # TODO: /Workspace/Users/<user>/project/notebooks/notebook_name
            source: # TODO: WORKSPACE
          environment_key: # TODO: Default
```

````python
# Task 3: Complete solution
solution = '''
resources:
  jobs:
    ingest_orders_job:
      name: ingest_orders_job
      tasks:
        - task_key: bronze_load
          notebook_task:
            notebook_path: /Workspace/Users/<user>/orders_pipeline/notebooks/bronze_load
            source: WORKSPACE
          environment_key: Default
'''
print('Task 3 solution:'); print(solution)
````

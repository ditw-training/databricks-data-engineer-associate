## Q20

Which statement correctly reads a JSON file with an explicit schema in PySpark?

**A.**
```python
df = spark.read.json("/path/data.json", schema="id INT, name STRING")
```

**B.**
```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType
schema = StructType([StructField("id", IntegerType()), StructField("name", StringType())])
df = spark.read.schema(schema).json("/path/data.json")
```

**C.**
```python
df = spark.read.format("json").option("schema", "id:int, name:string").load("/path")
```

**D.**
```python
df = spark.read.json("/path").cast("id INT, name STRING")
```

## Answer Key -- Day 1

| # | Ans | Explanation |
|---|-----|-------------|
| Q1 | **C** | Unity Catalog provides centralized access control, auditing, lineage, and data discovery across workspaces. |
| Q2 | **B** | The hierarchy is: Metastore > Catalog > Schema > Table/View/Function. |
| Q3 | **B** | Unity Catalog requires explicit `USE CATALOG` + `USE SCHEMA` grants before accessing objects within. |
| Q4 | **C** | `CREATE TABLE ... AS SELECT * FROM read_files()` (CTAS with `read_files`) is the standard pattern for loading files into managed Delta tables. |
| Q5 | **C** | Managed tables use catalog-managed storage; dropping them deletes the underlying data. External tables reference external locations and survive DROP. |
| Q6 | **A** | `createOrReplaceTempView()` registers a DataFrame as a session-scoped temporary view queryable via `spark.sql()`. |
| Q7 | **B** | The Delta transaction log (`_delta_log`) provides ACID guarantees, enables time travel, and enforces schema on write. |
| Q8 | **C** | Delta Lake stores data as Parquet files with an additional JSON-based transaction log. |
| Q9 | **B** | Each version in Delta History corresponds to a single transactional operation on the table. |
| Q10 | **B** | `VERSION AS OF` enables time travel, returning the table snapshot at the specified version number. |
| Q11 | **B** | Since late 2025, Serverless compute is the default compute type for new Databricks workspaces. |
| Q12 | **C** | Unity Catalog Volumes are the recommended replacement for DBFS root/mounts for storing non-tabular files. |
| Q13 | **A** | ELT (Extract-Load-Transform) loads raw data into the lakehouse first, then transforms in-place using Spark. ETL transforms before loading. Databricks favors ELT. |
| Q14 | **B** | `dropDuplicates(["col1", "col2"])` removes duplicate rows based on a subset of columns. `distinct()` works on all columns. |
| Q15 | **B** | Temporary views live only in the SparkSession (lost on cluster restart). Permanent views are persisted in Unity Catalog. |
| Q16 | **B** | `%run` executes another notebook in the same context, making its variables and functions available to the calling notebook. |
| Q17 | **B** | By default, Delta enforces schema -- the write fails. Use `mergeSchema=true` to enable schema evolution. |
| Q18 | **A** | `dbutils.fs` provides file system utilities: `ls()`, `cp()`, `rm()`, `head()`, `mkdirs()`. |
| Q19 | **B** | `RESTORE TABLE` reverts the table to a previous state and creates a new version entry in the history log. |
| Q20 | **B** | Define a `StructType` schema and pass it via `.schema(schema)` before calling `.json()`. |
